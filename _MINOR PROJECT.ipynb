```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
# Import Data
df = pd.read_csv("Maharashtra.csv")
```

```python
df['infected']=df['Confirmed']-(df['Cured']+df['Deaths']) # MAKING NEW COLUMN NAMED AS INFECTED
df['infected']
```

# CFR means Case Fatality Rate 
is the proportion of deaths from a certain disease compared to the total number of people diagnosed with the disease for a certain period of time.

```python
df['CFR']=(df['Deaths']/df['Confirmed'])*100
df['CFR']
```

# CMR means Crude Mortality Rate
is the number of deaths occurring among the population of a given geographical area

```python
pop=112374333 #approximate population of Maharashtra as in 2020
df['CMR']=(df['Deaths']/pop)*100000
df['CMR']
```

# IFR means Infection Fatality Rate
represents the proportion of deaths among all the infected individuals

```python
df['IFR']=(df['Deaths']/df['infected'])*100
df['IFR']
```

```python
#THE BELOW CASES ARE CASES PER DAY
df["new_deaths"]=df["Deaths"].diff(1)
df["new_cured"]=df["Cured"].diff(1)
df["new_confirmed"]=df["Confirmed"].diff(1)
```

```python
plt.plot(df['Date'], df['Confirmed'], label = "Confirmed")
plt.plot(df['Date'], df['Deaths'], label = "Deaths")
plt.plot(df['Date'], df['Cured'], label = "Cured")
plt.ylabel('CASES')

plt.title('GROWTH OF COVID 19 CASES IN MAHARASHTRA DURING JUNE 2020')
plt.xticks(df['Date'],rotation=90)
plt.legend(loc='best')
plt.show()
```

```python
plt.plot(df['Date'], df['new_confirmed'], label = "New Confirmed")
plt.plot(df['Date'], df['new_deaths'], label = "New Deaths")
plt.plot(df['Date'], df['new_cured'], label = "New Cured")
plt.ylabel('New Cases per day')
plt.title('COVID 19 CASES PER DAY IN MAHARASHTRA DURING JUNE 2020')
plt.xticks(df['Date'],rotation=90)
plt.legend(loc='best')
plt.show()
```

```python
df['new_CFR']=(df['new_deaths']/df['new_confirmed'])*100
df['new_CFR']
```

```python
df['new_CMR']=(df['new_deaths']/pop)*1000000
df['new_CMR']
```

```python
df['new_infected']=df['new_confirmed']-(df['new_cured']+df['new_deaths'])
df['new_infected']
```

```python
df['new_IFR']=(df['new_deaths']/df['new_infected'])*100
df['new_IFR']
```

```python
plt.plot(df['Date'], df['new_CFR'], label = "New CFR")
plt.plot(df['Date'], df['new_CMR'], label = "New CMR")
plt.plot(df['Date'], df['new_IFR'], label = "New IFR")
plt.ylabel('RATES')
plt.title('CFR/CMR/IFR PER DAY IN JUNE 2020')
plt.xticks(df['Date'],rotation=90)
plt.legend(loc='best')
plt.show()
```
