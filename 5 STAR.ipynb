```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
# Import Data
df1 = pd.read_csv("Maharashtra.csv")
df2 = pd.read_csv("gujarat.csv")
df3 = pd.read_csv("madhyapradesh.csv")
df4 = pd.read_csv("telangana.csv")
df5 = pd.read_csv("karnataka.csv")
df6 = pd.read_csv("goa.csv")
```

```python
df1['infected']=df1['Confirmed']-(df1['Cured']+df1['Deaths']) # MAKING NEW COLUMN NAMED AS INFECTED
df2['infected']=df2['Confirmed']-(df2['Cured']+df2['Deaths'])
df3['infected']=df3['Confirmed']-(df3['Cured']+df3['Deaths'])
df4['infected']=df4['Confirmed']-(df4['Cured']+df4['Deaths'])
df5['infected']=df5['Confirmed']-(df5['Cured']+df5['Deaths'])
df6['infected']=df6['Confirmed']-(df6['Cured']+df6['Deaths'])
```

```python
df1['CFR']=(df1['Deaths']/df1['Confirmed'])*100
df2['CFR']=(df2['Deaths']/df2['Confirmed'])*100
df3['CFR']=(df3['Deaths']/df3['Confirmed'])*100
df4['CFR']=(df4['Deaths']/df4['Confirmed'])*100
df5['CFR']=(df5['Deaths']/df5['Confirmed'])*100
df6['CFR']=(df6['Deaths']/df6['Confirmed'])*100
```

```python
df1['IFR']=(df1['Deaths']/df1['infected'])*100
df2['IFR']=(df2['Deaths']/df2['infected'])*100
df3['IFR']=(df3['Deaths']/df3['infected'])*100
df4['IFR']=(df4['Deaths']/df4['infected'])*100
df5['IFR']=(df5['Deaths']/df5['infected'])*100
df6['IFR']=(df6['Deaths']/df6['infected'])*100
```

```python
df1["new_deaths"]=df1["Deaths"].diff(1)
df1["new_cured"]=df1["Cured"].diff(1)
df1["new_confirmed"]=df1["Confirmed"].diff(1)

df2["new_deaths"]=df2["Deaths"].diff(1)
df2["new_cured"]=df2["Cured"].diff(1)
df2["new_confirmed"]=df2["Confirmed"].diff(1)

df3["new_deaths"]=df3["Deaths"].diff(1)
df3["new_cured"]=df3["Cured"].diff(1)
df3["new_confirmed"]=df3["Confirmed"].diff(1)

df4["new_deaths"]=df4["Deaths"].diff(1)
df4["new_cured"]=df4["Cured"].diff(1)
df4["new_confirmed"]=df4["Confirmed"].diff(1)

df5["new_deaths"]=df5["Deaths"].diff(1)
df5["new_cured"]=df5["Cured"].diff(1)
df5["new_confirmed"]=df5["Confirmed"].diff(1)

df6["new_deaths"]=df6["Deaths"].diff(1)
df6["new_cured"]=df6["Cured"].diff(1)
df6["new_confirmed"]=df6["Confirmed"].diff(1)

M=df1["new_confirmed"]
G=df2["new_confirmed"]
MP=df3["new_confirmed"]
T=df4["new_confirmed"]
K=df5["new_confirmed"]
GO=df6["new_confirmed"]
```

```python
plt.plot(df1['Date'], df1['new_confirmed'], label = "Maharashtra-Confirmed")
plt.plot(df2['Date'], df2['new_confirmed'], label = "Gujarat-Confirmed")
plt.plot(df3['Date'], df3['new_confirmed'], label = "MP-Confirmed")
plt.plot(df4['Date'], df4['new_confirmed'], label = "Telangana-Confirmed")
plt.plot(df5['Date'], df5['new_confirmed'], label = "Karnataka-Confirmed")
plt.plot(df6['Date'], df6['new_confirmed'], label = "Goa-Confirmed")
plt.ylabel('New Cases per day')
plt.title('COVID 19 confirmed CASES PER DAY DURING JUNE 2020')
plt.xticks(df1['Date'],rotation=90)
plt.legend(loc='best')
plt.show()
```
